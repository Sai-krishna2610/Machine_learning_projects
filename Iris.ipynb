# Define the problem
Given the measurements of sepal length,sepal width , petal length , petal width and the goal is to classify the iris flower into 3 species such as  setosa, versicolor, and virginica

Input:measurements of sepal length,sepal width , petal length , petal width

Output: predicted iris flower name

Success Criteria: Achieve an accuracy rate of at least 90% in correctly classifying the species of iris flowers.


# Data:
The dataset for the iris flower classfication can be import from sklearn or can be download from using this link - https://archive.ics.uci.edu/dataset/53/iris




```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

# Loading data

```python
iris=load_iris
```

```python
#converting the data  present in iris() to dataFrame by using pandas
df=pd.DataFrame(iris().data,columns=iris().feature_names)
```

```python
# Adding target Feature to dataFrame
df["target"]=iris().target
```

```python
from sklearn.model_selection import train_test_split
x=df.drop(['target','target_names'],axis='columns')
y=df.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
from sklearn.svm import SVC
model=SVC()#Support Vector Classifier
#If the hyperplane classifies the dataset linearly then the algorithm we call it as SVC
#And the algorithm that separates the dataset by non-linear approach then we call it as SVM.
model.fit(x_train,y_train)
print(f"accuracy score is {model.score(x_test,y_test)} ")
```

```python
from sklearn.svm import SVC
model=SVC(C=100)#C parameter in SVM is Penalty parameter of the error term.
model.fit(x_train,y_train)
print(f"accuracy score is {model.score(x_test,y_test)} ")
```

```python
model.predict(x_test)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
# Define parameter grid for hyperparameter tuning
param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
}

# Create SVM classifier
svm = SVC()

# Initialize GridSearchCV
grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy')

# Perform grid search on training data
grid_search.fit(x_train, y_train)

# Get best hyperparameters
best_params = grid_search.best_params_

# Initialize SVM with best hyperparameters
best_svm = SVC(**best_params)

# Train SVM on full training data with best hyperparameters
best_svm.fit(x_train, y_train)

# Predict on test data
y_pred = best_svm.predict(x_test)

# Calculate accuracy on test data
accuracy = accuracy_score(y_test, y_pred)

print("Best Hyperparameters:", best_params)
print("Test Accuracy:", accuracy)
```
